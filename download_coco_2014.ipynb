```python
import os
import zipfile
from pathlib import Path
import urllib.request

# Куда сохраняем датасет относительно корня проекта
DATA_ROOT = Path("dataset")
DATA_ROOT.mkdir(exist_ok=True)

# Подпапка для train2014 в стиле fast-neural-style (можно просто dataset/train2014)
COCO_TRAIN_DIR = DATA_ROOT / "train2014"

# Ссылки на COCO 2014 train images (официальный сайт)
COCO_TRAIN_URL = "http://images.cocodataset.org/zips/train2014.zip"

COCO_TRAIN_ZIP = DATA_ROOT / "train2014.zip"

print("DATA_ROOT:", DATA_ROOT.resolve())
print("Будем скачивать COCO train2014 в:", COCO_TRAIN_ZIP)
```

```python
def download_with_progress(url: str, dst: Path):
    if dst.exists():
        print(f"Файл уже существует: {dst}, пропускаю скачивание.")
        return

    def _progress(block_num, block_size, total_size):
        downloaded = block_num * block_size
        percent = downloaded / total_size * 100 if total_size > 0 else 0
        print(f"\rСкачано: {downloaded / (1024*1024):.2f} MB ({percent:.1f}%)", end="")

    print(f"Скачиваю {url} -> {dst} ...")
    urllib.request.urlretrieve(url, dst, _progress)
    print("\nСкачивание завершено.")
```

```python
download_with_progress(COCO_TRAIN_URL, COCO_TRAIN_ZIP)
```

```python
if not COCO_TRAIN_DIR.exists():
    COCO_TRAIN_DIR.mkdir(parents=True, exist_ok=True)

print(f"Распаковываю {COCO_TRAIN_ZIP} в {COCO_TRAIN_DIR} ...")

with zipfile.ZipFile(COCO_TRAIN_ZIP, 'r') as zf:
    zf.extractall(DATA_ROOT)

print("Распаковка завершена.")
```
